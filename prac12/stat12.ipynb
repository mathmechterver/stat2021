# 12. Марковские цепи. Part II.

### Теорема. (Можно доказать по индукции [ref](https://stepik.org/lesson/254734/step/9?unit=234942))

Марковскую цепь также можно задать через след выражение и оно эквивалентно определению выше
$P(\xi_0 = x_0, \xi_1 = x_1,\ldots, \xi_n = x_n) = P(\xi_0=x_0)P(\xi_1=x_1)\ldots P(\xi_n=x_n)$

### Однородные цепи маркова

Марковская цепь является однородной, если вероятность перехода за несколько шагов из состояния $n$ в состояние $m$ зависит только от $n-m$

$P_{n, n+1}(x,y) = P(x,y) = T(x,y)$ - матрица перехода


### Стационарное распределение

$\pi^*$ называется стационарным, если для любого начального распределения $\pi$:
$$\lim_{n\to \infty} \pi * T^n = \pi^*$$

1. (3) **Модель популяционной генетики** [ref](https://stepik.org/lesson/254734/step/8?unit=234942)

Имеется $m$ диплоидных особей (т.е. у особи две копии каждого гена). Всего в генофонде $2m$ генов. Будем считать, что

* Число особей остается постоянным из поколения в поколение.
* Гены для любой особи в $(n+1)$ -м поколении выбираются случайным образом из генов в $n$-м поколении.

Рассмотрим один из генов, пусть у него есть два аллеля, обозначим их через $A$ и $B$. Пусть $\xi_n$ случайная величина равная количеству аллелей типа $A$ во всем генофонде. Тогда $\xi_n$ будет цепью Маркова с фазовым пространством $\{0,1,2,\dots,2m\}$. 

* (1) Найдите вероятность перехода $P_n(j,k)$.
* (1) Найдите условное математическое ожидание $\mathbb{E}(\xi_n\,|\,\xi_{n-1}=j)$. Исходя из него установите чему равно математическое ожидание $\mathbb{E}\xi_n=\sum\limits_{k=0}^{2m} kq_n(j,k)$, если в начальный момент времени количество аллелей типа $A$ равно $j$.
* (1) Найдите все возвратные состояния и поймите, что все невозвратные состояния являются сообщающимися. Тогда $\lim\limits_{n\to\infty} q_n(a,b)=0$ для всех невозвратных состояний $a$ и $b$. С помощью предыдущей задачи найдите предел $\lim\limits_{n\to\infty} q_n(j,2m)$. Тем самым будет установлена вероятность того, что при изначальном наличии в популяции $j$ аллелей $A$ и $2m-j$ аллелей $B$ вторые когда-нибудь полностью исчезнут.
