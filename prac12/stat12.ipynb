# 12. Марковские цепи. Марковский случайный процесс.

![title](markov.png)

$X$ - конечное или счетное множество (фазовое пространство $\in \mathcal{N}$).
Система принимает состояния из $X$, $\xi_n$ - ее состояние в момент времени $n$ ($\xi_n: \Omega \to X$)
Для любых $n \leq 1$ и $x_1, \ldots, x_{n+1} \in X$

$$P(\xi_{n+1}=x_{n+1}|\xi_0=x_0, \xi_1=x_1, \ldots, \xi_n=x_n) = P(\xi_{n+1}=x_{n+1})$$

$$P(\xi_0=x_0, \xi_1=x_1, \ldots, \xi_n=x_n) > 0$$


Вероятность перехода $P_n(x,y)$

1. $\forall n \in \mathbb{N}$ $\forall x \in X$ $\sum_{y \in X} P_n(x,y) = 1$
2. $\forall n \in \mathbb{N}$ $\forall x,y \in X$ $P_n(x,y) \leq 0$
3. $P_n(x,y) = P(\xi_n = y|\xi_{n-1} = x)$, если $P(\xi_{n-1}=x)>0$ 

Теорема.

$P(\xi_0 = x_0, \xi_1 = x_1,\ldots, \xi_n = x_n) = P(\xi_0=x_0)P(\xi_1=x_1)\ldots P(\xi_n=x_n)$

Вероятность перехода за несколько шагов

$P_{n,m}(x,y) = \sum_{x_{n+1}, \ldots x_{m-1}}P_{n+1}(x,x_{n+1})P_{n+2}(x_{n+1},x_{n+2})\ldots P_{m-1}(x_{m-1},y)$



### Однородные цепи маркова

$P_{n,m}$ зависит лишь от $n-m$
$P_{n, n+1}(x,y) = P_n(x,y) = Q(x,y)$ - матрица перехода

### Теорема

$Q_n = Q^n$

### Возвратные состаяния

$f_a(n) = P(\xi_n = a, \xi_{n-1} \neq a, \ldots, \xi_{1} \neq a|\xi_{0} = a)$ - вероятность возврата

* Если $\sum_{n=1}^{\infty}f_a(n)=1$, то $a$ - возвратное
* Если $\sum_{n=1}^{\infty}f_a(n)<1$, то $a$ - невозвратное
