# 7. Доверительные интервалы для параметров нормального распределения

## Теория 
[Доверительные интервалы](https://nsu.ru/mmf/tvims/chernova/ms/lec/node38.html)

Пусть $\hat \theta(X_{[n]})$ - некоторая оценка параметра $\theta$

Известная статистика $Y(\hat \theta, \theta)$, такая что 
* $F_Y$ функция распределения не зависит от $\theta$
* $Y$ - непрерывна и строго монотонная, а значит имеет обратную функцию $Y^{-1}(x)$

Тогда

$P\{y_{\alpha/2} < Y < y_{1-\alpha/2}\} = F(y_{1-\alpha/2}) - F(y_{\alpha/2}) = 1 - \alpha$

Получаем доверительный интервал

$P\{Y^{-1}(y_{\alpha/2}) < \theta < Y^{-1}(y_{1-\alpha/2})\} = F(y_{1-\alpha/2}) - F(y_{\alpha/2}) = 1 - \alpha$


--------------------------------
### Пример для нормального распределения 



Оценка матожидания: $\overline{X} = \frac{\sum_{i=1}^n X_i}{n} \sim \mathcal{N}(\mu, \frac{\sigma}{n})$

Статистика: $Y = \sqrt{n}\frac{\overline{X} - \mu}{\sigma} \sim \mathcal{N}(0, 1)$

$1-\alpha$-интервал для матожидания

$(1-\alpha) = P\Big(u_{\frac{\alpha}{2}} < \sqrt{n}\frac{\overline{X} - \mu}{\sigma} < u_{1 - \frac{\alpha}{2}}\Big) = P\Big(\overline{X} - u_{\frac{\alpha}{2}}\frac{\sigma}{\sqrt{n}} > \mu > \overline{X} - u_{1 - \frac{\alpha}{2}}\frac{\sigma}{\sqrt{n}}\Big)$

----------------

### Работает только в случае если знаем $\mu$ и $\sigma$, а что если нет?

Пусть есть просто выборка из произвольного нормального распределения

$X_{[n]}$ из $\xi \in \mathcal{N}(a, \sigma)$

Точечные оценки:

$\overline{X} = \frac{1}{n}\sum_{i=1}^{N}X_i$

$s^2 = \frac{1}{n-1}\sum_{i=1}^{n}(X_i - \overline{X})^2$

$Y_{1} = (n-1)\frac{s^2}{\sigma^2} = \sum_{i=1}^{n}\Big(\frac{X_i - \overline{X}}{\sigma}\Big)^2 \sim \chi^2(n-1)$

$Y_{2} = \frac{\overline{X} - a}{s / \sqrt{n}} = \frac{\sqrt{n}\frac{\overline{X} - a}{\sigma}}{s/\sigma} \sim T(n-1)$


Построение доверительных интервалов с доверительной вероятностью $(1-\alpha)$

* Для матожидания 
$$P\Big(\overline{X} - \frac{s}{\sqrt{n}}t_{1-\frac{\alpha}{2}} < a < \overline{X} + \frac{s}{\sqrt{n}}t_{1-\frac{\alpha}{2}}\Big) = 1 - \alpha$$


* Для дисперсии 
$$P\Big(\frac{(n-1)s^2}{\chi^2_{1 - \alpha / 2}}< \sigma^2 < \frac{(n-1)s^2}{\chi^2_{\alpha / 2}}\Big) = 1 - \alpha$$

### Мразь 1

Пусть $ z_{1}, \ldots ,z_{k}$ — совместно независимые стандартные нормальные случайные величины, то есть: $ z_{i}\sim \mathcal{N}(0,1)$. Тогда случайная величина

$$X=z_{1}^{2}+\ldots +z_{k}^{2}$$
имеет распределение хи-квадрат с $k$ степенями свободы, то есть $$X \sim \chi ^{2}(k)$$

### Мразь 2

Пусть $Y_{0},Y_{1},\ldots ,Y_{n}$ — независимые стандартные нормальные случайные величины, такие что $ Y_{i}\sim {\mathcal {N}}(0,1),\;i=0,\ldots ,n$. Тогда распределение случайной величины $t$, где

$T={\frac {Y_{0}}{\sqrt {{\frac {1}{n}}\sum \limits _{i=1}^{n}Y_{i}^{2}}}}$,
называется распределением Стьюдента с $n$ степенями свободы $t\sim \mathrm {T}(n)$
