# 13. Пуассоновский процесс. Непрерывные марковские цепи.

$(\Omega, \mathcal{F}, \mathbb{P})$ - вероятностное пространство, T - мн-во моментов времени

$\{\xi_t\}_{t\in T}$ - семейство случайных величин

$\xi_t(\omega) = f(t, \omega)$

### Пуассоновский процесс

$T = [0, +\infty)$ $\xi_t$ - число событий на промежутке $[0,t]$
1. За конечное время происходит лишь конечное число событий
2. Появление нескольких событий одновременно - невозможно
3. На непересекающихся промежутках количества событий - независимы
$\xi_{t_1}, \xi_{t_2} - \xi_{t_1}, \ldots, \xi_{t_n} - \xi_{t_{n-1}}$
4. Количество зарегестрированных событий зависит от только длины промежутка ($\xi_{t+h} - \xi_t$ не зависит от $t$)

1. $\lim_{h \rightarrow 0}\frac{1}{h}P(\xi_{t+h} - \xi_{t} > 1) = 0$
[Доказательство](https://stepik.org/lesson/254732/step/3?unit=234940)

2. $P(\xi_{t+h} - \xi_t > 0) = 1-e^{-\lambda h} = \lambda h + o(h)$
[Доказательство](https://stepik.org/lesson/254732/step/4?unit=234940)

3. $P(\xi_t = 0) = \frac{(\lambda t)^k}{k!}e^{-\lambda t}$ 
[Доказательство](https://stepik.org/lesson/254732/step/5?unit=234940)

5. Рассмотрим пуассоновский процесс с интенсивностью $\lambda$. Пусть случайная величина $\tau_k$ обозначает момент времени, когда была зарегистрирована $k$-я частица. Тогда
$$P(\tau_1>t)=P(\xi_t=0)=\frac{(\lambda t)^0}{0!} e^{-\lambda t}=e^{-\lambda t}$$
Следовательно, 
$$F_{\tau_1}(t)=P(\tau_1\leqslant t)=1-P(\tau_1>t)=1-e^{-\lambda t} $$
Таким образом, $\tau_1$ имеет экспоненциальное распределение с параметром $\lambda$.
Найдите математическое ожидание и дисперсию $\tau_k$
