# 8. Гипотезы. P-value. Проверка простых гипотез.

## Теория
[Все о гипотезах](http://www.sevin.ru/fundecology/literature/hypot.html)

В статистике мы как-то оцениваем параметры распределений, но ничего не можем утверждать *точно*. Однако мы можем узнать с какой вероятностью выполняется то или иное утверждение по данным.  

Например с какой вероятностью "$\xi$ распределена по закону $F_0$".

В этом случае пользуются механизмом **проверок статистических гипотез**.

* $H_0: \xi \sim F_0$

Гипотезы бывают 2-х типов:
* Простая гипотеза (однозначно задано распределение с параметрами): $\mu = 5$, $\sigma = 0.1$
* Сложная гипотеза (известен класс распределения): $\xi \sim \mathcal{N}(\mu, \sigma)$, $\mu \in [0,1]$



------------

## Проверке гипотез на простом примере

Пусть у нас есть некоторая выборка $X_{[n]}$ из $\xi$ и мы хотим проверить гипотезу, что выборка взята из некоторого распределения с известным параметром $\theta$. 

Вводим нулевую гипотезу $H_0: \hat\theta = \theta$

Пусть есть оценка $\hat\theta(X_{[n]}) \sim F(x|H_0)$. 

**ВАЖНО: мы рассматриваем распределение НЕ $\xi$, а ОЦЕНКИ ПАРАМЕТРА $\hat\theta$**

Мы хотим узнать действительно ли данная оценка корректна на наших данных.

Для этого нам нужно понять, попадает ли оценка в **критическую область** (маловероятная область $k$)

![title](types_of_hypothesis.png)

## Альтернативая гипотеза

По рисунку сверху видно, что возможны несколько вариантов критических областей.
Этот выбор зависит от **альтернативной гипотезы**.

Пусть мы рассматриваем простую гипотезу о том, что наша оценка с равна некоторому значению $\hat\theta = \theta_0$

В зависимости от задачи можно выбрать альтернативную гипотезу:
* $\hat\theta > \theta_0$
* $\hat\theta < \theta_0$
* $\hat\theta \neq \theta_0$

Альтернативная гипотеза $H_1$ **принимается** , если **отклоняется** $H_0$. (Оценка попала в критическую область) 

## P-value
Есть 2 равнозначных подхода проверить гипотезу
* Найти **критическое значение** ($k$ на картинке) и сравнить со статистикой 
* Найти **p-value** (вероятность того, что статистика критерия примет значение такое же как $z(X_{[n]})$ или более *экстремальное*)

![title](pvalue.jpeg)

### Статистики:

ЦПТ
$$Y_0 = \sqrt{n}\frac{\overline{X} - \mu}{\sigma} \sim \mathcal{N}(0, 1)$$

Для дисперсии при неизвестном матожидании (критерий Стьюдента)
$$Y_{1} = (n-1)\frac{s^2}{\sigma^2} = \sum_{i=1}^{n}\Big(\frac{X_i - \overline{X}}{\sigma}\Big)^2 \sim \chi^2(n-1)$$

Для матожидания при неизвестной дисперсии (критерий хи-квадрат)
$$Y_{2} = \frac{\overline{X} - a}{s / \sqrt{n}} = \frac{\sqrt{n}\frac{\overline{X} - a}{\sigma}}{s/\sigma} \sim T(n-1)$$
