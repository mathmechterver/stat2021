# 8. Гипотезы. P-value.

Есть задача, когда нам нужно проверить некоторое "утверждение" на валидность. Например $\xi$ распределена по закону $F_0$ или $F_1$.

* $H_0: \xi \sim F_0$
* $H_1: \xi \sim F_1$

В этом случае пользуются механизмом проверок статистических гипотез.

* Простая гипотеза (однозначно задано распределение с параметрами): $\mu = 5$, $\sigma = 0.1$
* Сложная гипотеза (известен класс распределения): $\xi \sim \mathcal{N}(\mu, \sigma)$, $\mu \in [0,1]$



Наша задача: принять решение об истинности нулевой гипотезы $H_0$ против альтернативной гипотезы $H_1$

При использовании критериев значимости в качестве нулевой гипотезу выбирают ту, которая ближе всего к простой гипотезе.

------------

## Проверке гипотез

Пусть статистика $Z \sim F_Z(\cdot|H_0)$

$z(X_{[n]})$

$p-value$ - статистика, равная вероятности того, что статистика критерия примет значение такое же как $z(X_{[n]})$ или более *экстремальное* при условии, что верна нулевая гипотеза


### Схема проверки статистических p-value

1. Формулируем нулевую и альтернативную гипотезу
2. Выбираем уровень значимости ($\alpha$)
3. Выбираем критерий и статистику критерия ($Z$)
4. Вычисляем выборочное знаение статистики и критерия ($Z_B = Z(X_{[n]})$)
5. Вычисляем p-value
6. Сравниваем p-value с уровнем значимости
7. Принимаем статистическое решение
