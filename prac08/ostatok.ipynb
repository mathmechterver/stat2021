# Схема проверки статистических гипотез

1. Формулируем нулевую и альтернативную гипотезу
2. Выбираем уровень значимости
3. Выбираем критерий и статистику критерия
4. Строим критическую область при условии справедливости нулевой гипотезы
5. Вычисляем выборочное знаение статистики
6. Проверяем, принадлежит ли выборочное значение статистики критической области
7. Принимаем статистическое решение 

**Статистический критерий значимости** (критерий значимости) -- правило проверки стат. гипотезы при некоторой фиксированной альтернативе

**Статистика критерия значимости** - статистика, по значениям которой судят о справедливости гипотезы.

**Уровень значимости** $\alpha$ называется столь малая вероятноть, что событие с такой вероятностью является практически невозможным.

**Критическая область критерия значимости** - подобласть значений статистики критерия, вероятность попадания в которую для этой статистики при условии истинности нулевой гипотезы равна уровню значимости.
