2. Пусть $\xi\sim N(a,\sigma)$, $a$ неизвестно, $\sigma=2$

Проверяется гипотеза $H_0:a=5$, при альтернативе $H_1=5.5$
Какого объема следует взять выборку, чтобы вероятность ошибки первого рода не превосходила  0.05, а вероятность ошибки второго рода не превосходила 0.1?  Найдите наименьшее значение

Ответ: ближайшее наименьшее целое,  большее полученного значения

Формат ответа: значение объема выборки
